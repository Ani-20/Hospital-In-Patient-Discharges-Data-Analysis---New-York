```python
# Step 1: Create a DataFrame with non-NaN values for 'Zip_Code_-_3_digits'
filtered_df = df.dropna(subset=['Zip_Code_-_3_digits'])

# Step 2: Create a dictionary to map 'Zip_Code_-_3_digits' to ('Health_Service_Area', 'Hospital_County')
zip_mapping = filtered_df.groupby('Zip_Code_-_3_digits')[['Health_Service_Area', 'Hospital_County']].first().to_dict(orient='index')

# Step 3: Fill missing 'Health_Service_Area' and 'Hospital_County' values based on the mapping
def fill_missing_values(row):
    zip_code = row['Zip_Code_-_3_digits']
    if pd.notna(zip_code) and zip_code in zip_mapping:
        mapping = zip_mapping[zip_code]
        row['Health_Service_Area'] = mapping['Health_Service_Area']
        row['Hospital_County'] = mapping['Hospital_County']
    return row

df = df.apply(fill_missing_values, axis=1)

# Display the updated DataFrame
display(df)
```

```python
# Iterate through rows of the DataFrame
for index, row in df.iterrows():
    zip_code = row['Zip_Code_-_3_digits']
    health_service_area = row['Health_Service_Area']
    hospital_county = row['Hospital_County']
    
    # Check if 'Zip_Code_-_3_digits' has a value and 'Health_Service_Area' is "None"
    if pd.notna(zip_code) and health_service_area == "None":
        # Find a matching row with the same 'Zip_Code_-_3_digits'
        matching_row = df[df['Zip_Code_-_3_digits'] == zip_code].iloc[0]
        
        # Get the values from the matching row and update the current row
        df.at[index, 'Health_Service_Area'] = matching_row['Health_Service_Area']
        df.at[index, 'Hospital_County'] = matching_row['Hospital_County']

# Display the updated DataFrame
display(df)
```
